# Libraries to Import

```python
import pandas as pd
import numpy as np
import os
```

# Read In Processed Coin Datasheet

```python
parent_folder = os.path.dirname(os.path.dirname(os.getcwd())) #Parent folder of the repo
model_folder = "models"
model_data_file_path = os.path.join(parent_folder, model_folder) #path to models folder

#load in data from cvs files
coin_df_wmd = pd.read_csv(open(os.path.join(model_data_file_path, "coin-similarity-matrix-description.csv"), "r"), index_col="Name")
coin_df_ed = pd.read_csv(open(os.path.join(model_data_file_path, "coin-similarity-matrix-euclidean-distance.csv"), "r"), index_col="Name")

#test output of the processed wmd data file.
coin_df_wmd
```

```python
#test output of the processed euclidean distance data file.
coin_df_ed
```

# Create and Display Weighted Average Matrix

```python
coin_df_average = coin_df_ed.add(coin_df_wmd)/2
coin_df_average
```

We want to cluster coins using volatility, by using the euclidean distance measure, coin description, by using the word mover's distance, and by using an average of both of those measures. 

# Sanity Check

```python
#extracting iota wmd and euclidean distance similarity value for iota and anchor-protocol
wmd = coin_df_wmd["iota"]["anchor-protocol"]
ed = coin_df_ed["iota"]["anchor-protocol"]

average = (wmd + ed)/2

assert average == coin_df_average["iota"]["anchor-protocol"], "Coins iota and anchor-protocol fail unit test. Computed word mover's distances do not match."
print("Coins iota and anchor-protocol pass the unit test. They have a euclidean distance of: " + str(average))
```

# Download Similarity Matrix as a CSV File

```python
coin_df_average.to_csv(open(os.path.join(model_data_file_path, "coin-similarity-matrix-weighted-average.csv"), "w"))
```
